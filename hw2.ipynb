```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import networkx as nx
```

```python
#데이터 불러오기
```

```python
data = pd.read_csv('C:\googleplaystore.csv', encoding= 'unicode_escape')
```

```python
g = nx.Graph()
g = nx.nx.from_pandas_edgelist(data, source = 'Genres', target = 'Category') #장르가 다르더라도 같은 카테고리로 묶이므로 연관성 테스트
print(nx.info(g))
#Node정보 출력
```

```python
nx.degree_centrality(g) # Node 중심성 측정(얼마나 많이 연결되어 있는가)
```

```python
nx.eigenvector_centrality(g) #고유벡터의 중심성 구하기(중요성 있는 것과 얼마나 연결되어 있는가)
```

```python
#종류별 연관 그래프
```

```python
pos = nx.spring_layout(g)
betCent = nx.betweenness_centrality(g, normalized=True, endpoints=True)
node_color = [20000.0 * g.degree(v) for v in g]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(g, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
# 중간 중심성으로 차수와 노드 크기에 따라 색상이 달라지게 함
plt.axis('off');
#출력
```

```python
plt.figure(figsize=(30, 30))
pos = nx.spring_layout(g, k = 0.30)
nx.draw_networkx(g,pos, node_size = 25, node_color = 'blue')
plt.show()
# 그래프 그리기
```
